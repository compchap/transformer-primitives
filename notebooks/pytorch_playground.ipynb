```python
import time
import asyncio

async def say_after(delay, what):
    await asyncio.sleep(delay)
    print(what)

async def main():
    print(f"started at {time.strftime('%X')}")

    await say_after(1, 'hello')
    await say_after(2, 'world')

    print(f"finished at {time.strftime('%X')}")

# asyncio.run(main())
await main()
```

```python
import tiktoken
import torch
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = "Your journey starts with one step"
```

```python
integers = tokenizer.encode(text)
```

```python
integers = torch.tensor(integers)
```

```python
integers
```

```python
torch.manual_seed(123)
vocab_size = 50257
output_dim = 3
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
```

```python
probs = torch.softmax(embedding_layer.weight, dim=1)
```

```python
print(embedding_layer(integers))
```

```python
def softmax_naive(x):
    expFn = torch.exp(x) / torch.exp(x).sum(dim=0)
    return expFn / expFn.sum(dim=0)
```

```python
embedding_layer_normalized = softmax_naive(embedding_layer(integers))
```

```python
embedding_layer_normalized
```
